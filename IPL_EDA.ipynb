# ***Who has the most 'player of the match' trophy?***

```python
#grouping all the players who got trophy n sorting
playerofmatches_df = ipl_df.groupby('player_of_match')[['id']].count()
playerofmatches_df = playerofmatches_df.sort_values('id',ascending = True).reset_index()
playerofmatches_df.rename(columns={'id':'total_trophies','player_of_match':'player_name'} , inplace =True)
```

```python
top_player = playerofmatches_df.tail(15)
top_player
```

```python
plt.figure(figsize=(12,8))
plt.title('Top 15 palyers awarded Player Of Match trophy',fontsize = 20,fontweight = 'bold')
plt.xlabel('Highest Number of Trophies',fontsize = 20,fontweight = 'bold')
plt.ylabel('Players name',fontsize = 20,fontweight = 'bold')
plt.tick_params(labelsize = 12)
plt.barh(top_player.player_name,top_player.total_trophies,color = 'firebrick');
```

# Which decission has proved right maximun times?

```python
balling = ipl_df.loc[(ipl_df['toss_winner'] == ipl_df['winner']) & (ipl_df['toss_decision'] == 'field'),['id','winner','toss_decision']]
```

```python
a = balling.winner.count()
a
```

```python
bat = ipl_df.loc[(ipl_df['toss_winner'] == ipl_df['winner']) & (ipl_df['toss_decision'] == 'bat'),['id','winner','toss_decision']]
```

```python
b = bat.winner.count()
b
```

```python
result = pd.concat([bat,balling])
result = result.groupby('toss_decision')[['id']].count().reset_index()
result.rename(columns ={'toss_decision':'Decission','id':'Total'},inplace = True)
```

```python
result
```

```python
plt.figure(figsize=(8,8))
plt.grid()
plt.title('Toss Decision Proved Right',fontsize=25,fontweight='bold')
plt.xlabel('Decision',fontsize=20,fontweight='bold')
plt.ylabel('Total',fontsize=20,fontweight='bold')
plt.tick_params(labelsize=20)
plt.bar(decision.Decision,decision.Total , color = ['coral'])
plt.bar(result.Decission , result.Total, color = ['royalblue']) 
plt.legend(['Decission taken','Decission proved right']);
```

# *Which team has won ipl trophy most of times?*

```python
final = ipl_df.groupby('season').tail(1)
final
```

```python
final.winner.unique()
```

```python
final_df=final['winner'].value_counts().reset_index()
final_df.rename(columns={'index':'team','winner':'total'},inplace = True)
final_df
```

```python
plt.figure(figsize=(10,10))
plt.title("Season Winners",fontweight='bold',fontsize=35)
plt.xlabel('Teams',fontweight='bold',fontsize=30)
plt.ylabel('Total Seasons',fontweight='bold',fontsize=30)
plt.xticks(rotation='58')
plt.tick_params(labelsize=15)
plt.bar(final_df.team,final_df.total,color = ['blue','yellow','darkorange','purple','cornflowerblue']);
```
